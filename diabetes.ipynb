```python
# basic imports
import numpy as np
import pandas as pd

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# for data visualization  for EDA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('diabetes.csv')
```

```python
#snapshot of data
data.head()
#data
```

```python
#show information of the data set - column names and types
data.info()
```

```python
# loop through the columns and check the missing values
for col in data.columns:
    missing_data = data[col].isnull().mean()
    print(f'{col} - {missing_data: .1%}')
```

```python
#summary statistics of the data - round the figures to 2 decimal places 
data.describe().applymap('{:,.2f}'.format)
```

```python
# Data exploration to answer the question raised
```

```python
# 1.How many females from the data have no records of pregnancy?
# In the pregnancies column, a value of zero indicates no pregnancy 
print(data['Pregnancies'].eq(0).sum())
#111 females from the records show no pregnancy
```

```python
#2. What is the average blood pressure for the records that have never been pregnant ?
zero_pregnancy = data[['Pregnancies','BloodPressure']][(data["Pregnancies"] == 0)]
zero_pregnancy
```

```python
#Average blood pressure for females with no pregnancy recorded
print(zero_pregnancy['BloodPressure'].mean().round(2))
```

```python
#3. What is the average blood pressure for the records that have 1 or more pregancies ?
yes_pregnancy = data[['Pregnancies','BloodPressure']][(data["Pregnancies"] > 0)]
yes_pregnancy
```

```python
#Average blood pressure for females with one or more pregnancy recorded  is 69.44
print(yes_pregnancy['BloodPressure'].mean().round(2))
```

```python
#4. What is the average Glucose level  for the records that have 1 or more pregancies ?
glucose_pregnancy = data[['Pregnancies','Glucose']][(data["Pregnancies"] > 0)]
glucose_pregnancy
```

```python
#Average glucose level for females with one or more pregnancy recorded  is 120.54
print(glucose_pregnancy['Glucose'].mean().round(2))
```
