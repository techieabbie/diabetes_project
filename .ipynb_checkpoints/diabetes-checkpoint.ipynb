```python
# basic imports
import numpy as np
import pandas as pd

# for data visualization  for EDA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('diabetes.csv')
```

```python
#snapshot of data
data.head()
#data
```

```python
#show information of the data set - column names and types
data.info()
```

```python
# loop through the columns and check the missing values
for col in data.columns:
    missing_data = data[col].isnull().mean()
    print(f'{col} - {missing_data: .1%}')
```

```python
#summary statistics of the data - round the figures to 2 decimal places 
data.describe().applymap('{:,.2f}'.format)
```

```python
# Data exploration to answer the question raised
```

```python
# 1.How many females from the data have no records of pregnancy?
# In the pregnancies column, a value of zero indicates no pregnancy 
print(data['Pregnancies'].eq(0).sum())
#111 females from the records show no pregnancy
```

```python
#2.Out of the females that shows no pregnancies, how many tested positive for diabetes?
#38 females have who have records of no pregancies tested positive for diabetes.
print ((data[['Pregnancies','Outcome']][(data["Pregnancies"] == 0) & (data["Outcome"] == 1)]).count())
```

```python
#2. What is the average blood pressure for the profile that have never been pregnant but tested positive?
zero_pregnancy = data[['Pregnancies','Outcome','BloodPressure']][(data["Pregnancies"] == 0) & (data["Outcome"] == 1)]
zero_pregnancy
```

```python
#Average blood pressure for Females with no pregnancy history who do not have diabetes is 63.11
print(zero_pregnancy['BloodPressure'].mean().round(2))
```
